```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = {
    'Year': [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Cell Phone Usage (%)': [10, 15, 20, 25, 30, 40, 50, 60, 65, 70, 75, 80, 85, 90, 92, 93, 95, 96, 97, 98, 99]
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.lineplot(x='Cell Phone Usage (%)', y='year', data=df, marker='o', color='b', linewidth=2)
plt.title('Rise in Cell Phone Usage Over the Years', fontsize=16)
plt.xlabel('Cell Phone Usage (%)', fontsize=12)
plt.ylabel('year', fontsize=12)
plt.grid(True)
plt.show()
```
