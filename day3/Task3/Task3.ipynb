```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019],
    'USA': [7.3, 7.4, 7.6, 7.8, 7.9, 8.1, 8.2, 8.3, 8.4, 8.5],
    'Canada': [7.5, 7.6, 7.7, 7.8, 7.9, 8.0, 8.1, 8.3, 8.4, 8.5],
    'Australia': [7.2, 7.3, 7.4, 7.5, 7.7, 7.8, 7.9, 8.0, 8.1, 8.3]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))  
plt.plot(df['Year'], df['USA'], marker='o', color='b', label='USA', linestyle='-', linewidth=2, markersize=6)
plt.plot(df['Year'], df['Canada'], marker='s', color='g', label='Canada', linestyle='-', linewidth=2, markersize=6)
plt.plot(df['Year'], df['Australia'], marker='^', color='r', label='Australia', linestyle='-', linewidth=2, markersize=6)
plt.title('Happiness Scores Over Years', fontsize=16)
plt.xlabel('Happiness Score', fontsize=14)
plt.ylabel('year', fontsize=14)
plt.legend(title='Countries')
plt.grid(True)
plt.show()
```
