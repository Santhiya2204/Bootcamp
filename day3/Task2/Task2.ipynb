```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Income (USA)': [7700, 7800, 7900, 79000, 79000],
    'Income (Canada)': [7900, 8000, 81000, 8200, 82500],
    'Income (Germany)': [7800, 79000, 8000, 81000, 8100]
}
df = pd.DataFrame(data)
df.set_index('Year', inplace=True)
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, palette="tab10", linewidth=2)
plt.title('income  Over Time by Country')
plt.xlabel('income (Years)')
plt.ylabel('Year')
plt.legend(title='Country')
plt.show()
```
