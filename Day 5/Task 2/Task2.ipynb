```python
import matplotlib.pyplot as plt
import numpy as np
years = np.arange(1990, 2025, 5)
total_cell_phones = [0.01, 0.09, 0.74, 1.16, 2.00, 3.50, 4.60, 6.00]
cell_phones_per_person = [0.002, 0.015, 0.12, 0.18, 0.31, 0.50, 0.60, 0.75]
fig, ax1 = plt.subplots()
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Cell Phones (Billions)', color='tab:blue')
ax1.plot(years, total_cell_phones, marker='o', linestyle='-', color='tab:blue', label='Total Cell Phones')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax2 = ax1.twinx()
ax2.set_ylabel('Cell Phones per Person', color='tab:red')
ax2.plot(years, cell_phones_per_person, marker='s', linestyle='--', color='tab:red', label='Cell Phones per Person')
ax2.tick_params(axis='y', labelcolor='tab:red')
plt.title('Rise of Cell Phones Over Time')
ax1.grid(True, linestyle='--', alpha=0.6)

fig.tight_layout()
plt.show()
```
